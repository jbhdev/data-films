```python
# Importer les bibliothèques nécessaires
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict

# Charger les données
df = pd.read_csv('streamlit_app/datasets/raw/films.csv')

# Fonction pour convertir la chaîne de genres en liste
def parse_genres(genre_str):
    try:
        if pd.isna(genre_str):
            return []
        genres = ast.literal_eval(genre_str)
        return genres if isinstance(genres, list) else []
    except:
        return []

# Appliquer la fonction à la colonne des genres
df['genres_list'] = df['genres'].apply(parse_genres)

# Créer des dictionnaires pour stocker les sommes et les comptes
genre_sums = defaultdict(float)
genre_counts = defaultdict(int)

# Parcourir chaque film et ses genres
for _, row in df.dropna(subset=['vote_average', 'genres_list']).iterrows():
    for genre in row['genres_list']:
        genre_sums[genre] += row['vote_average']
        genre_counts[genre] += 1

# Calculer les moyennes
genre_avg = {genre: (genre_sums[genre] / count) for genre, count in genre_counts.items()}

# Créer un DataFrame avec les résultats
genre_ratings = pd.DataFrame({
    'Genre': list(genre_avg.keys()),
    'Note moyenne': list(genre_avg.values()),
    'Nombre de films': [genre_counts[g] for g in genre_avg.keys()]
})

# Filtrer les genres avec peu de films (optionnel)
min_films = 10  # Nombre minimum de films pour qu'un genre soit inclus
genre_ratings = genre_ratings[genre_ratings['Nombre de films'] >= min_films]

# Trier par note moyenne et prendre les 15 premiers
top_genres = genre_ratings.sort_values('Note moyenne', ascending=False).head(15)

# Créer la figure
plt.figure(figsize=(14, 10))
sns.set_style("whitegrid")

# Créer le graphique à barres
ax = sns.barplot(x='Note moyenne', y='Genre', data=top_genres, 
                 palette='viridis', hue='Genre', legend=False)

# Ajouter les titres et les labels
plt.title('Top 15 des genres les mieux notés', fontsize=18, pad=20)
plt.xlabel('Note moyenne', fontsize=14, labelpad=10)
plt.ylabel('Genre', fontsize=14, labelpad=10)

# Ajouter les valeurs sur les barres
for i, (_, row) in enumerate(top_genres.iterrows()):
    ax.text(row['Note moyenne'] + 0.02, i, 
           f"{row['Note moyenne']:.2f} ({row['Nombre de films']} films)", 
           va='center', fontsize=10)

# Ajuster les marges
plt.tight_layout()

# Afficher les statistiques
print("=== Statistiques des genres ===")
print(f"Nombre total de genres analysés : {len(genre_ratings)}")
print(f"Note moyenne globale : {df['vote_average'].mean():.2f}")

# Afficher le graphique
plt.show()

# Afficher le tableau des résultats
print("\nDétails des 15 genres les mieux notés :")
display(top_genres[['Genre', 'Note moyenne', 'Nombre de films']]
       .style.background_gradient(subset=['Note moyenne'], cmap='YlGn')
       .format({'Note moyenne': '{:.2f}'}))
```

```python
# Importer les bibliothèques nécessaires
import pandas as pd
import ast
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Charger les données
df = pd.read_csv('streamlit_app/datasets/raw/films.csv')

# Préparer les données pour l'analyse des pays de production
country_list = []
for countries in df['production_countries'].dropna():
    try:
        # Convertir la chaîne en liste Python
        countries = ast.literal_eval(countries)
        country_list.extend(countries)
    except:
        continue

# Compter les occurrences de chaque pays
country_counts = pd.Series(country_list).value_counts()

# Prendre les 15 premiers pays et regrouper le reste dans "Autres"
top_countries = country_counts.head(15)
other_count = country_counts[15:].sum()
top_countries['Autres'] = other_count

# Créer la figure
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Créer l'histogramme horizontal
ax = sns.barplot(x=top_countries.values, y=top_countries.index, palette="viridis")

# Ajouter les titres et les labels
plt.title('Répartition des pays de production', fontsize=18, pad=20)
plt.xlabel('Nombre de films', fontsize=14, labelpad=10)
plt.ylabel('Pays', fontsize=14, labelpad=10)

# Ajouter les valeurs sur les barres
for i, v in enumerate(top_countries.values):
    ax.text(v + 5, i, f'{v:,}', color='black', va='center', fontweight='bold', fontsize=10)

# Améliorer la lisibilité
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Ajouter une grille pour une meilleure lisibilité
ax.xaxis.grid(True, linestyle='--', alpha=0.7)

# Afficher les statistiques
print(f"Nombre total de pays uniques : {len(country_counts)}")
print(f"Nombre total de films analysés : {len(df)}")
print("\nTop 15 des pays de production :")
for country, count in top_countries.items():
    print(f"{country}: {count} films")

# Ajuster les marges
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
# Importer les bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Charger les données
df = pd.read_csv('streamlit_app/datasets/raw/films.csv')

# Convertir la colonne release_date en datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Obtenir la date d'aujourd'hui
today = pd.Timestamp(datetime.now().date())

# Créer une colonne pour le statut de sortie
df['status_sortie'] = df['release_date'].apply(
    lambda x: 'Déjà sorti' if pd.notnull(x) and x <= today else 'Pas encore sorti'
)

# Compter les films par statut
status_counts = df['status_sortie'].value_counts()

# Créer la figure
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Créer le graphique à barres
ax = sns.barplot(x=status_counts.index, y=status_counts.values, palette='viridis')

# Ajouter les titres et les labels
plt.title('Répartition des films par statut de sortie', fontsize=16, pad=20)
plt.xlabel('Statut de sortie', fontsize=12, labelpad=10)
plt.ylabel('Nombre de films', fontsize=12, labelpad=10)

# Ajouter les valeurs sur les barres
for i, v in enumerate(status_counts.values):
    ax.text(i, v + 5, f'{v:,}', color='black', ha='center', fontweight='bold')

# Afficher les statistiques
total_films = len(df)
print(f"Nombre total de films analysés : {total_films:,}")
print(f"Films déjà sortis : {status_counts.get('Déjà sorti', 0):,} ({(status_counts.get('Déjà sorti', 0)/total_films)*100:.1f}%)")
print(f"Films à venir : {status_counts.get('Pas encore sorti', 0):,} ({(status_counts.get('Pas encore sorti', 0)/total_films)*100:.1f}%)")

# Ajuster les marges
plt.tight_layout()

# Afficher le graphique
plt.show()
```

```python
# Importer les bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Charger les données
df = pd.read_csv('streamlit_app/datasets/raw/films.csv')

# Convertir la colonne release_date en datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# Extraire l'année de sortie
df['release_year'] = df['release_date'].dt.year

# Supprimer les lignes où l'année n'est pas disponible
df = df.dropna(subset=['release_year'])

# Créer des décennies (arrondir à la dizaine inférieure)
df['decade'] = (df['release_year'] // 10) * 10

# Compter les films par décennie
decade_counts = df['decade'].value_counts().sort_index()

# Créer la figure
plt.figure(figsize=(14, 8))
sns.set_style("whitegrid")

# Créer le graphique à barres
ax = sns.barplot(x=decade_counts.index, y=decade_counts.values, palette="viridis")

# Personnaliser les étiquettes de l'axe X pour afficher les décennies
plt.xticks(rotation=45, ha='right')
ax.set_xticklabels([f"{int(x)}-{int(x)+9}" for x in decade_counts.index])

# Ajouter les titres et les labels
plt.title('Répartition des films par décennie', fontsize=18, pad=20)
plt.xlabel('Décennie', fontsize=14, labelpad=10)
plt.ylabel('Nombre de films', fontsize=14, labelpad=10)

# Ajouter les valeurs sur les barres
for i, v in enumerate(decade_counts.values):
    ax.text(i, v + 5, f'{v:,}', color='black', ha='center', fontweight='bold')

# Ajuster les marges
plt.tight_layout()

# Afficher les statistiques
print(f"Période couverte : {int(df['release_year'].min())} - {int(df['release_year'].max())}")
print(f"Nombre total de films analysés : {len(df):,}")
print(f"Nombre de décennies représentées : {len(decade_counts)}")

# Afficher le graphique
plt.show()
```

```python
# Importer les bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.ticker import FuncFormatter



# Charger les données
df = pd.read_csv('streamlit_app/datasets/raw/films.csv')

# Paramètres ajustables
percentile_popularite = 85  # Pourcentage de popularité à considérer
top_n = 15  # Nombre de films à afficher dans le top
min_annee = 1950  # Année minimale à considérer

# Nettoyage des données
df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
df = df[df['release_year'] >= min_annee]

# Calcul du seuil de popularité
seuil_popularite = df['popularity'].quantile(percentile_popularite/100)

# Création de la figure
fig = plt.figure(figsize=(18, 30))
fig.suptitle('Analyse des Films', fontsize=20, y=1.02)

# 1. Distribution des notes moyennes
plt.subplot(4, 2, 1)
sns.histplot(df['vote_average'].dropna(), bins=20, kde=True, color='#3498db')
plt.title('Distribution des notes moyennes', fontsize=14)
plt.xlabel('Note moyenne', fontsize=12)
plt.ylabel('Nombre de films', fontsize=12)

# 2. Distribution de la popularité (échelle logarithmique)
plt.subplot(4, 2, 2)
sns.histplot(np.log10(df['popularity'].replace(0, 0.1)), bins=20, kde=True, color='#e74c3c')
plt.title('Distribution de la popularité (échelle logarithmique)', fontsize=14)
plt.xlabel('log10(Popularité)', fontsize=12)
plt.ylabel('Nombre de films', fontsize=12)

# 3. Top des films les mieux notés
plt.subplot(4, 2, (3, 4))
top_movies = df[df['popularity'] >= seuil_popularite][['original_title', 'vote_average', 'popularity', 'release_year']].copy()
top_movies = top_movies.sort_values('vote_average', ascending=False).head(top_n)

ax = sns.barplot(x='vote_average', y='original_title', data=top_movies, 
                 palette='viridis', hue='release_year', dodge=False)
plt.title(f'Top {top_n} des films les mieux notés\n(Popularité ≥ {percentile_popularite}e percentile)', fontsize=14)
plt.xlabel('Note moyenne', fontsize=12)
plt.ylabel('')
plt.legend(title='Année', bbox_to_anchor=(1.05, 1), loc='upper left')

# 4. Évolution des notes moyennes par année
plt.subplot(4, 2, 5)
df_yearly = df.groupby('release_year')['vote_average'].mean().reset_index()
sns.lineplot(x='release_year', y='vote_average', data=df_yearly, marker='o', color='#2ecc71')
plt.title('Évolution de la note moyenne par année', fontsize=14)
plt.xlabel('Année de sortie', fontsize=12)
plt.ylabel('Note moyenne', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# 5. Durée des films par année
plt.subplot(4, 2, 6)
if 'runtimeMinutes' in df.columns:
    df_yearly_runtime = df.groupby('release_year')['runtimeMinutes'].mean().reset_index()
    sns.lineplot(x='release_year', y='runtimeMinutes', data=df_yearly_runtime, 
                 color='#9b59b6', marker='o')
    plt.title('Durée moyenne des films par année', fontsize=14)
    plt.xlabel('Année de sortie', fontsize=12)
    plt.ylabel('Durée (minutes)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)

# 6. Top des langues originales
plt.subplot(4, 2, 7)
top_languages = df['original_language'].value_counts().head(10)
sns.barplot(y=top_languages.index, x=top_languages.values, palette='rocket')
plt.title('Top 10 des langues originales', fontsize=14)
plt.xlabel('Nombre de films', fontsize=12)
plt.ylabel('Langue', fontsize=12)

# 7. Distribution par décennie
plt.subplot(4, 2, 8)
if 'decennie' in df.columns:
    decade_counts = df['decennie'].value_counts().sort_index()
    sns.barplot(x=decade_counts.index, y=decade_counts.values, palette='coolwarm')
    plt.title('Distribution des films par décennie', fontsize=14)
    plt.xlabel('Décennie', fontsize=12)
    plt.ylabel('Nombre de films', fontsize=12)
    plt.xticks(rotation=45)

# Ajuster l'espacement
plt.tight_layout()

# Afficher les statistiques
print("=== Statistiques descriptives ===")
print(f"Nombre total de films : {len(df):,}")
print(f"Période : {int(df['release_year'].min())} - {int(df['release_year'].max())}")
print(f"Note moyenne : {df['vote_average'].mean():.2f} ± {df['vote_average'].std():.2f}")
print(f"Popularité médiane : {df['popularity'].median():.2f}")
print(f"Seuil de popularité ({percentile_popularite}e percentile) : {seuil_popularite:.2f}")

# Afficher les graphiques
plt.show()

# Afficher le tableau des meilleurs films
print(f"\n=== Top {top_n} des films (parmi les {percentile_popularite}% les plus populaires) ===")
display(top_movies[['original_title', 'vote_average', 'popularity', 'release_year']]
        .rename(columns={
            'original_title': 'Titre',
            'vote_average': 'Note',
            'popularity': 'Popularité',
            'release_year': 'Année'
        }).style.background_gradient(subset=['Note', 'Popularité'], cmap='YlGnBu'))
```

```python
# Importer les bibliothèques nécessaires
import pandas as pd
import ast  # Pour convertir les chaînes en listes
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import numpy as np

# Charger les données
df = pd.read_csv('streamlit_app/datasets/raw/films.csv')

# Convertir la colonne release_date en année
df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year

# Fonction pour convertir la chaîne de genres en liste
def parse_genres(genre_str):
    try:
        if pd.isna(genre_str):
            return []
        # Essayer de convertir la chaîne en liste
        genres = ast.literal_eval(genre_str)
        # S'assurer que c'est bien une liste
        return genres if isinstance(genres, list) else []
    except:
        return []

# Appliquer la fonction à la colonne des genres
df['genres_list'] = df['genres'].apply(parse_genres)

# Compter les occurrences de chaque genre
genre_counter = Counter()
for genres in df['genres_list']:
    genre_counter.update(genres)

# Créer un DataFrame à partir du compteur
genre_df = pd.DataFrame(genre_counter.most_common(), columns=['Genre', 'Nombre de films'])
genre_df['Pourcentage'] = (genre_df['Nombre de films'] / len(df)) * 100

# Créer une figure avec 2 sous-graphiques
plt.figure(figsize=(18, 30))
sns.set_style("whitegrid")

# 1. Graphique à barres des genres les plus courants
plt.subplot(2, 1, 1)
ax = sns.barplot(x='Nombre de films', y='Genre', 
                 data=genre_df.head(20),  # Afficher les 20 premiers genres
                 palette='viridis')

# Ajouter les valeurs sur les barres
for i, v in enumerate(genre_df['Nombre de films'].head(20)):
    ax.text(v + 5, i, f"{v:,} ({genre_df['Pourcentage'].iloc[i]:.1f}%)", 
            color='black', va='center')

plt.title('Top 20 des genres les plus représentés', fontsize=16, pad=20)
plt.xlabel('Nombre de films', fontsize=12)
plt.ylabel('Genre', fontsize=12)



# Créer un DataFrame avec une ligne par film et par genre
genre_yearly = []
for _, row in df[df['release_year'].notna()].iterrows():
    for genre in row['genres_list']:
        genre_yearly.append({
            'Année': int(row['release_year']),
            'Genre': genre
        })

genre_yearly_df = pd.DataFrame(genre_yearly)

# Calculer le nombre de films par genre et par année
genre_yearly_counts = genre_yearly_df.groupby(['Année', 'Genre']).size().reset_index(name='Nombre')

# Garder seulement les 10 genres les plus courants
top_genres = genre_df['Genre'].head(10).tolist()
genre_yearly_top = genre_yearly_counts[genre_yearly_counts['Genre'].isin(top_genres)]

# Pivoter pour avoir les genres en colonnes
genre_pivot = genre_yearly_top.pivot(index='Année', columns='Genre', values='Nombre').fillna(0)



# Ajuster l'espacement
plt.tight_layout()

# Afficher les statistiques
print("=== Statistiques des genres ===")
print(f"Nombre total de genres uniques : {len(genre_df)}")
print(f"Nombre moyen de genres par film : {df['genres_list'].apply(len).mean():.2f}")
print("\nTop 10 des genres :")
display(genre_df.head(10))

# Afficher les graphiques
plt.show()
```
